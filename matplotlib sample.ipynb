```python
import matplotlib.pyplot as plt
import numpy as np

# 기본 설정 3x3 grid 그림판 사이즈 8x8
fig, axes = plt.subplots(nrows=3, ncols=3, 
                         sharex=True, sharey=True,
                         figsize=(12,6)
                         )
x = range(5)
y = range(5)
print(x)
print("grid shape :" , axes.shape)
print("pixel size :",fig)

for row in axes:
    for col in row:
        col.plot(x, y)
    
    #col caption
for ax, col in zip(axes[0,:], ['1', '2', '3']):
    ax.set_title(col, size=20)
    
    #row caption
for ax, row in zip(axes[:,0], ['A', 'B', 'C']):
    ax.set_ylabel(row, size=20, rotation=0, labelpad=15)
  
plt.show()
```

```python
import matplotlib.pyplot as plt

x = range(10)
y = range(10)

fig, ax = plt.subplots(nrows=2,ncols=2)

for row in ax:
    for col in row:
        col.plot(x, y)
        col.set_title('title')  
        col.set_xlabel('x-axis')
        col.set_ylabel('y-axis')
                
fig.tight_layout()
            
plt.show()
```

```python
import matplotlib.pyplot as plt

x = range(10)
y = range(10)

fig, axes = plt.subplots(nrows=3,ncols=3)

for cnt, ax in enumerate(axes.ravel()):
    if cnt < 7:    
        ax.plot(x, y)
    else:
        ax.axis('off') # hide subplot
plt.show()
```
